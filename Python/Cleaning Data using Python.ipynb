## Data Cleaning

**Data cleaning** is a crucial step in the data analysis process that ensures the accuracy, consistency, and reliability of the data being used. 

It typically involves such operations as:
- Removing duplicates
- Dropping unused columns
- Removing whitespace characters (spaces, tabs, and newlines) and other symbols from column values
- Splitting columns into several ones
- Fixing inconsistencies in data
- Removing rows based on certain conditions
- Remove or replace null values

### Project Goal
In this portfolio project, I'll showcase my data cleansing skills by cleaning the `"Customer Call List.xlsx"` dataset using the Pandas library.

```python
# Import pandas library
import pandas as pd
```

```python
# Import the dataset excel file to Pandas dataframe and view its contents
df = pd.read_excel("Customer Call List.xlsx")
df
```

```python
# Drop duplicate entries (rows 19 and 20 are duplicated)
df = df.drop_duplicates()
df
```

```python
# Drop unused column 'Not_Useful_Column'
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# Clean the 'Last_Name' column as it contains unnecessary characters, '/', '...', and '_'
df["Last_Name"] = df["Last_Name"].str.strip("/._")
df
```

```python
# Clean the 'Phone_Number' column values and bring them to the unified format "xxx-xxx-xxxx" through a number of steps

# Remove all symbols from phone numbers except digits
df["Phone_Number"] = df["Phone_Number"].str.replace('[^0-9]','')

# Convert 'Phone_Number' column values to string data type to unlock further transformation opportunities
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

# Bring 'Phone_Number' values to unified format "xxx-xxx-xxxx"
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

# Remove 'nan--' values from phone numbers
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

# Remove '--' values from phone numbers
df["Phone_Number"] = df["Phone_Number"].str.replace('--','')

df
```

```python
# Split 'Address' column into three separate ones, 'Street_Address', 'State', and 'Zip_Code', and add them to the dataframe

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', 2, expand=True)
df
```

```python
# Now, 'Paying Customer' and 'Do_Not_Contact' columns have 'Yes', 'Y', 'No', 'N' distinct values. Remove the inconsistency
# so that the columns contain only 'Yes' and 'No' values

df['Paying Customer'] = df['Paying Customer'].replace({'Y': 'Yes', 'N': 'No'})
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Y': 'Yes', 'N': 'No'})
df
```

```python
# Replace 'N/a' and null values in the dataframe with empty values, ''.

df = df.replace('N/a','')
df=df.fillna('')
df
```

```python
# Delete entries with customers who don't want to be contacted ('Do_Not_Contact' = 'Yes')

df = df[df['Do_Not_Contact'] != 'Yes']
df
```

```python
# Update index so that after cleaning the dataframe it does not have missing values
df = df.reset_index(drop=True)
df
```

```python
# Export the cleaned dataset to Excel

df.to_excel('cleaned_dataset.xlsx', index=False)
```
